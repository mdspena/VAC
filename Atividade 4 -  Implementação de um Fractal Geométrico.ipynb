### Atividade 4 -  Implementação de um Fractal Geométrico ###

```python
import turtle
```

```python
import time

minimo = 1

def build_tree(t, comprimento, angulo):
    if comprimento > minimo:
        t.forward(comprimento)
        novo_comprimento = comprimento / 2
        build_tree(t, novo_comprimento, angulo)
        t.left(angulo)
        build_tree(t, novo_comprimento, angulo)
        t.right(angulo * 2)
        build_tree(t, novo_comprimento, angulo)
        t.left(angulo)
        t.backward(comprimento)

tree = turtle.Turtle()
tree.hideturtle()
tree.speed(200)

tree.screen.bgcolor('black')
starts = [90,210,330]
tree.color('white')

time.sleep(20)

for start in starts:
    tree.setheading(start)
    build_tree(tree, 140, 120)
turtle.mainloop()
```

**Outro fractal que fiz**

```python
minimo = 5
it = 0

def build_tree(t, comprimento, reducao, angulo, it):
    if comprimento > minimo:
        tree.color(colors[it])
        t.forward(comprimento)
        novo_comprimento = comprimento - reducao
        t.left(angulo)
        it = it+1
        build_tree(t, novo_comprimento, reducao, angulo, it)
        t.right(angulo * 2)
        build_tree(t, novo_comprimento, reducao, angulo, it)
        t.left(angulo)
        tree.color(colors[it-1])
        t.backward(comprimento)

tree = turtle.Turtle()
tree.hideturtle()
tree.speed(200)

tree.screen.bgcolor("black")
starts = [0,90,180,270]
colors = ['green','yellow','blue','green','yellow','blue','green','yellow','blue']

for start in starts:
    tree.setheading(start)
    build_tree(tree, 50, 5, 17.5, it)
turtle.mainloop()
```
